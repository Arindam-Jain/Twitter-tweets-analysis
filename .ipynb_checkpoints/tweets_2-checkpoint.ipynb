```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
print(train.shape)
```

```python
trainle=train
trainle.keyword=le.fit_transform(trainle.keyword)
```

```python
X=trainle[['keyword','text']].values
y=trainle.target.values
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
X=ohe.fit_transform(X)
X.toarray()
```

```python
import nltk
from nltk.corpus import stopwords
import string
```

```python
def text_process(text):
    nopunc = [char for char in text if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
train['text'].apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
nltk.download('stopwords')
```

```python
bow_transformer = CountVectorizer(analyzer=text_process).fit(train['text'])
print(len(bow_transformer.vocabulary_))
```

```python
text_bow = bow_transformer.transform(train['text'])
```

```python
print('Shape of Sparse Matrix: ', text_bow.shape)
print('Amount of Non-Zero occurences: ', text_bow.nnz)
```

```python
text4=train['text'][34]
bow4 = bow_transformer.transform([text4])
print(bow4)
print(bow4.shape)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer().fit(text_bow)
tfidf4 = tfidf_transformer.transform(bow4)
print(tfidf4)
```

```python
text_tfidf = tfidf_transformer.transform(text_bow)
print(text_tfidf.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
Target_detect_model = MultinomialNB().fit(text_tfidf, train['target'])
```

```python
all_predictions = Target_detect_model.predict(text_tfidf)
print(all_predictions)
```

```python
from sklearn.metrics import classification_report
print (classification_report(train['target'], all_predictions))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('bow', CountVectorizer(analyzer=text_process)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
pipeline.fit(X_train,y_train)
```
